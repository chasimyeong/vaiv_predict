```python
import os
import requests
import numpy as np

#해당url에 request하기 위한 함
def request_prediction(url, file_name, form, command):
    files = {'image' : open(file_name, 'rb')}
    
    data = {'format' : form, 
           'command' : command}
    
    return requests.post(url, files=files, data = data)
```

```python
# sample image
file_path = './images/samples'
url = 'http://localhost:5000/Dpredict'
```

```python
%%time
file_1 = os.path.join(file_path, 'sample (1).jpg')
form_1 = 'total_base64'
command_1 = 'skyline_detection'

#request후 json파일에서 output_img인 key의 value 호출 (string으로 된 base64형태의 이미지)
test = request_prediction(url, file_1, form_1, command_1)
output_base64 = test.json()['result']['output_img']
```

```python
import base64
from PIL import Image
from io import BytesIO
import matplotlib.pyplot as plt

#base64 포맷 Python에서 호출한 경우
img = Image.open(BytesIO(base64.b64decode(output_base64[2:-1].encode())))
img
```
