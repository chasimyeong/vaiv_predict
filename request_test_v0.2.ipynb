```python
import os
import requests
import numpy as np

#해당url에 request하기 위한 함
def request_prediction(url, file_name, form):
    files = {'image' : open(file_name, 'rb'),
            'format' : form}
    return requests.post(url, files=files)
```

```python
# sample image
file_path = './images/samples'
url = 'http://localhost:5000/skylinePredict'
```

```python
%%time
file = os.path.join(file_path, 'sample (1).jpg')
form = 'total_base64'

#request후 json파일에서 output_img인 key의 value 호출 (string으로 된 base64형태의 이미지)
test = request_prediction(url, file, form)
output_base64 = test.json()['output_img']
```
